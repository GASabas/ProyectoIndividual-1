```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
```

```python
# Cargar dataset
df_movies_final = pd.read_csv('movies_final.csv')
```

```python
# Información básica del dataset
print("Información del Dataset:")
print(df_movies_final.info())

print("\nPrimeras filas del Dataset:")
print(df_movies_final.head())

# Resumen de valores nulos
print("\nValores nulos por columna:")
print(df_movies_final.isnull().sum())
```

```python
# Estadistica descriptiva y tipos de datos 
# Tipos de datos
print("\nTipos de datos:")
print(df_movies_final.dtypes)

# Resumen estadístico de variables numéricas
print("\nEstadísticas descriptivas de variables numéricas:")
print(df_movies_final.describe())

# Resumen estadístico de variables categóricas
print("\nResumen de variables categóricas:")
cat_columns = df_movies_final.select_dtypes(include='object').columns
print(df_movies_final[cat_columns].describe())
```

```python
# Frecuencia de géneros
df_movies_final['genre_names'].value_counts().plot(kind='bar', figsize=(10, 6), color='orange')
plt.title("Frecuencia de Géneros")
plt.xlabel("Género")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#Analizamos correlaciones
#  Seleccionar solo columnas numéricas
numeric_df = df_movies_final.select_dtypes(include=['float64', 'int64'])

# Verificar el DataFrame numérico resultante
print("Columnas numéricas seleccionadas para el heatmap:")
print(numeric_df.head())

# Heatmap de correlaciones
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Mapa de Calor de Correlaciones")
plt.show()
```

```python
# Identificamos patrones
#  Ingresos promedio por género
avg_revenue_by_genre = df_movies_final.groupby('genre_names')['revenue'].mean().sort_values(ascending=False)
print("\nIngresos promedio por género:")
print(avg_revenue_by_genre)

# Visualización de ingresos promedio por género
avg_revenue_by_genre.plot(kind='bar', figsize=(10, 6), color='purple')
plt.title("Ingresos Promedio por Género")
plt.xlabel("Género")
plt.ylabel("Ingresos Promedio")
plt.show()

# Calificaciones promedio por género
avg_popularity_by_genre = df_movies_final.groupby('genre_names')['popularity'].mean().sort_values(ascending=False)
print("\nCalificaciones promedio por género:")
print(avg_popularity_by_genre)

# Visualización de calificaciones promedio por género
avg_popularity_by_genre.plot(kind='bar', figsize=(10, 6), color='teal')
plt.title("Calificaciones Promedio por Género")
plt.xlabel("Género")
plt.ylabel("Calificación Promedio")
plt.show()
```
